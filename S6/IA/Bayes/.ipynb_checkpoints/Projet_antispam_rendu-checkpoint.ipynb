### Nettoyage de la base de données

```python
input_file = "spam.csv"
default = pd.read_csv(input_file)
default.head()
```

```python
base = default.dropna()
```

```python
cpt_ham = 0
cpt_spam = 0
mot = set()
for i in range (base.size // 4):
    t = base.iloc[i]
    for w in (re.findall(r"[\w]+",base.iloc[i][2])):
        mot.add(w)
    if t[3] == 0:
        cpt_ham+=1
    else:
        cpt_spam+=1
x = ['ham','spam']
y = [cpt_ham,cpt_spam]

plt.bar(x,y, width = 0.6)
plt.show()
print("Nombres de mots différents :", len(mot))
```

## Construction du dictionnaire de mots

Dictionnaire de fréquence

```python
def construction_dico_frequence(X,taille_mot):

dict_mot_frequence = dict()

for i in range(len(X)):
    
    #check if spam
    if default.iloc[i][3] == 1:
        #retire les points, virgules, et autres caracteres de mise en page
        mot = re.findall(r"[\w]+",default.iloc[i][2])
        for w in mot:
            if len(w) > taille_mot:
                value = dict_mot_frequence.get(w, None)
                if value is not None: 
                    dict_mot_frequence[w] = value+1
                else:
                    dict_mot_frequence[w] = 1
```

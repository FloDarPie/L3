```python
# on cherche a determiner le minimum d'une fonction
var('x')
f = x*x - 2*x + 1
u=5
plot(f, -u,u)
```

```python
def grad1(f, origine, pas, difference, nbIteration):
    
    g = f.diff(x)
    i = 1
    
    xk = f(x=origine)
    xk1 = xk - pas*g(x=xk)
    L=[xk]
    
    while i != nbIteration and abs(xk - xk1) > difference  :
        L.append(xk1)
        print(i,":",xk)
        xk = xk1
        xk1 = xk - pas*g(x=xk)
        i+=1
    return L

L = grad1(f,2.5,0.1,0.001,100)
```

```python
s = [(i,f(x=i)) for i in L]
u=s[0][0]
plot(f, 0,u,color='indigo') + point(s,pointsize=50,color='mediumseagreen')
```

```python
f = x*x - 1/4*x*x*x*x
u=5
plot(f, -u,u)
```

```python
#recherche de minimum
grad1Dessin(f,1.5,0.1,0.001,100)

#recherche de maximum
grad1Dessin(f,1.5,-0.1,0.001,100)
```

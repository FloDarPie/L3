```python
var('x' , 'y')
f(x,y) = x*x*x + 3*x*y*y - 15 *x -12 *y

u = 1.2
a = 2.2
plot3d(f, (-a,a), (-u,u))

```

```python
print(f.diff(x))
print(f.diff(y))

gradient = f.gradient()

```

```python
for x in range(-5,5):
    for y in range(-5,5):
        if gradient(x,y) == 0 :
            print(x,y)
```

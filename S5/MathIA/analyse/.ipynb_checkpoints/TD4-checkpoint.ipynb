```python
f(x, y) = x*x + x*y - y*y*y


pretty_print(diff(f, x))
pretty_print(diff(f, y))
v = f.gradient()
pretty_print(v)
```

```python
u = 5
H = plot_vector_field(
    v , (x,-u,u) , (y,-u,u)
)
show(H)
plot3d(f, (x,-u,u) , (y,-u,u))
```

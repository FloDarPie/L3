```python
#taylor a l'ordre 1 pour une fonction a 2 variables
def taylor1(f, A):
    if len(A)!= 2 :
        return "le second paramètre doit être une coordonnées de R^2"
    return f(A[0],A[1]) + f.diff(x)(A[0],A[1]) * (x -A[0]) + f.diff(y)(A[0],A[1]) * (y -A[1])

show(taylor1(f,A))
show(taylor1(f,B))
show(taylor1(f,[3]))
```

```python
#appliquer Taylor d'une fonction f en un point A et sur un point B proche de A
def applicationTaylor1(f, a, b):
    show("le gradiant de la fonction est :",f.gradient())
    t = taylor1(f,a)
    print("Taylor en :",a)
    print("(x,y) |-->", t)
    print("Pour",b)
    return t(x=b[0],y=b[1])
```

```python
A = [0,1]
B = [0.005, 0.995]
show(applicationTaylor1(f,A,B))
```

```python
def planTangent(f,A,u):
    pf = plot3d(f,(-u, u) ,(-u ,u), color='peru')
    zA = taylor1(f,A)
    pzA = plot3d(zA,(-u, u) ,(-u ,u), color = 'palevioletred')
    show(pf+pzA)
```

```python
planTangent(f,A,u)
```
